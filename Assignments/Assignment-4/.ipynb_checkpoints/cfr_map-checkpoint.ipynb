```python
#Matt McColeman 100525235
%pylab inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation
import math
import urllib
from mpl_toolkits.basemap import Basemap
import csv

Category=[]
Lat=[]
Long=[]
Cases=[]

with open('Source_data_for_CFR_vaccine_map_abridged.csv') as file:
    reader = csv.DictReader(file)
    for line in reader:
        Category.append(line['Category'])
        Lat.append((float)(line['Lat']))
        Long.append((float)(line['Long']))
        Cases.append((float)(line['Cases']))
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))

n=0

plt.figure(figsize=(10, 14))

earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for i in range (n, len(Category)):
    if (Category[n] == 'Measles'):
        x, y = earth(Long[n], Lat[n])

        size=math.log((Cases[n]/105000)*140+5)*5
        trans=0.5-Cases[n]/200000
        
        earth.plot(x, y, 'ro',markersize=size,alpha=trans)
    n+=1
    
plt.title("Measles outbreaks (100525235)")
plt.show()

n=0

plt.figure(figsize=(10, 14))

earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for i in range (n, len(Category)):
    if (Category[n] == 'Rubella'):
        x, y = earth(Long[n], Lat[n])
        
        size=math.log((Cases[n]/105000)*140+5)*5
        trans=0.5-Cases[n]/200000
        
        earth.plot(x, y, 'mo',markersize=size,alpha=trans)
    n+=1
    
plt.title("Rubella outbreaks (100525235)")
plt.show()

n=0

plt.figure(figsize=(10, 14))

earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for i in range (n, len(Category)):
    if (Category[n] == 'Mumps'):
        x, y = earth(Long[n], Lat[n])
        
        size=math.log((Cases[n]/105000)*140+5)*5
        trans=0.5-Cases[n]/200000
        
        earth.plot(x, y, 'go',markersize=size,alpha=trans)
    n+=1
    
plt.title("Mumps outbreaks (100525235)")
plt.show()

n=0

plt.figure(figsize=(10, 14))

earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for i in range (n, len(Category)):
    if (Category[n] == 'Polio'):
        x, y = earth(Long[n], Lat[n])
        
        size=math.log((Cases[n]/105000)*140+5)*5
        trans=0.5-Cases[n]/200000
        
        earth.plot(x, y, 'bo',markersize=size,alpha=trans)
    n+=1
    
plt.title("Polio outbreaks (100525235)")
plt.show()
```
